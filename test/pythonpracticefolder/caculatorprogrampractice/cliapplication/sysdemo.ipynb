```python
import json

movie_lists = [
    {"name": "Inception", "year": 2010,"cast":"B"},
    {"name": "Interstellar", "year": 2014,"cast":"A"},
    {"name":"RRR","year":2022,"cast":"A"}
]

with open("movie_lists.json", "w") as f:
    json.dump(movie_lists, f, indent=4)

print(json.dumps(movie_lists, indent=4))
```
