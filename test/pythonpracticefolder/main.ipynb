```python
###def function_name(parameter)
def is_prime(number)->bool:
    is_prime_result = True
    if number < 2:
        is_prime_result = False
    else:
        index = 2
        while index < number:
            if number % index == 0:
                is_prime_result = False
                break
            index = index + 1

    return is_prime_result
```

```python
is_prime(7)
```

```python
is_prime(8)
```

```python
print(is_prime(-7))
```
